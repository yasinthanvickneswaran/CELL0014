```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def sdot2(s,t,params):
    ML, PL, MT, PT = s
    km0L, kmL, KL, nL, kdmL, kpL, kdpL, km0T, kmT, KT, nT, kdmT, kpT, kdpT = params

    rate_ML_prod = km0L + kmL*(KT**nT/(PT**nT+KT**nT))
    rate_ML_loss = kdmL*ML

    rate_PL_prod =  kpL*ML
    rate_PL_loss =  kdpL*PL

    rate_MT_prod = km0T + kmT*(KL**nL/(PL**nL+KL**nL))
    rate_MT_loss = kdmT*MT

    rate_PT_prod = kpT*MT
    rate_PT_loss = kdpT*PT
        
    dML = rate_ML_prod - rate_ML_loss
    dPL = rate_PL_prod - rate_PL_loss
    
    dMT = rate_MT_prod - rate_MT_loss
    dPT = rate_PT_prod - rate_PT_loss
    
    ds = [dML, dPL, dMT, dPT]
    return ds

# define parameter values
km0L = 0.01
kmL = 5
KL = 500
nL = 2
kdmL = 0.1386 
kpL = 1.2
kdpL = 0.0165

km0T = 0.01
kmT = 5
KT = 500
nT = 2
kdmT = 0.1386 
kpT = 1.2
kdpT = 0.0165

params = [km0L, kmL, KL, nL, kdmL, kpL, kdpL,
          km0T, kmT, KT, nT, kdmT, kpT, kdpT]

# define initial conditions
ML0 = 0
PL0 = 0
MT0 = 36
PT0 = 2600

s0 = [ML0, PL0, MT0, PT0]

# define time observation points
t_start = 0
t_end = 1000

t_obs = np.arange(t_start, t_end+0.1, 1)

# run simulation
s_obs = odeint(sdot2, s0, t_obs, args=(params,))

# extract the observations
ML_obs = s_obs[:,0]
PL_obs0 = s_obs[:,1]
MT_obs = s_obs[:,2]
PT_obs0 = s_obs[:,3]
```

```python
def sdot3(s, t, params):
    ML, PL, MT, PT = s
    km0L, kmL, KL, n, kdmL, kpL, kdpL, km0T, kmT, KT, kdmT, kpT, kdpT = params

    rate_ML_prod = km0L + kmL * (KT**n / (PT**n + KT**n))
    rate_ML_loss = kdmL * ML
    rate_PL_prod = kpL * ML
    rate_PL_loss = kdpL * PL

    rate_MT_prod = km0T + kmT * (KL**n / (PL**n + KL**n))
    rate_MT_loss = kdmT * MT
    rate_PT_prod = kpT * MT
    rate_PT_loss = kdpT * PT

    dML = rate_ML_prod - rate_ML_loss
    dPL = rate_PL_prod - rate_PL_loss
    dMT = rate_MT_prod - rate_MT_loss
    dPT = rate_PT_prod - rate_PT_loss
            
    return [dML, dPL, dMT, dPT]

Pl_final1 = []
Pl_final2 = []

km0L = 0.01
kmL = 5
KL = 500
kdmL = 0.1386 
kpL = 1.2
kdpL = 0.0165
km0T = 0.01
kmT = 5
KT = 500    
kdmT = 0.1386 
kpT = 1.2
kdpT = 0.0165

n_vals = np.arange(1, 4.1, 0.1)

for n in n_vals:
    params = [km0L, kmL, KL, n, kdmL, kpL, kdpL, km0T, kmT, KT, kdmT, kpT, kdpT]

    # Initial conditions for Tetr High Initial
    ML0 = 36
    PL0 = 2600
    MT0 = 0
    PT0 = 0
    s0 = [ML0, PL0, MT0, PT0]

    t_start = 0
    t_end = 10000
    t_obs = np.arange(t_start, t_end + 0.1, 1)
    s_obs = odeint(sdot3, s0, t_obs, args=(params,))
    Pl_final1.append(s_obs[-1, 1])  # Append final PL value for Tetr High Initial

    # Initial conditions for Tetr Low Initial
    ML0 = 0
    PL0 = 0
    MT0 = 36
    PT0 = 2600
    s0 = [ML0, PL0, MT0, PT0]

    t_obs = np.arange(t_start, t_end + 0.1, 1)
    s_obs1 = odeint(sdot3, s0, t_obs, args=(params,))
    Pl_final2.append(s_obs1[-1, 1])  # Append final PL value for Tetr Low Initial

# Plotting
plt.figure(figsize = (12, 8))
plt.plot(n_vals, Pl_final1, label="Tetr High Initial")
plt.plot(n_vals, Pl_final2, label="Tetr Low Initial")
plt.xlabel("n")
plt.ylabel("Final Lacl Protein")
plt.title("Behaviour of Toggle Switch System at different n values")
plt.legend()
plt.show()
```

```python
# template code for the 2D bifurcation plot analysis

# plot 2c on Toggle Switch paper shows results across
# a range of kmL and kmT values on log scales. 
# we can use np.logspace() to create a set of 
# values that will be equally spaced on a log plot

# start with num_vals = 10
# so 10x10 grid points will be plotted
# when code is working you can increase (e.g. to 40)
# to cover more gridpoints
num_vals = 40

# create kmL and kmT values ranging
# from 10^0 to 10^3, i.e. spanning from 1 to 1000
kmL_vals = np.logspace(0,3,num_vals)
kmT_vals = np.logspace(0,3,num_vals)

# make figure that the results will be plotted into
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot()
ax.set_xlabel("kmL")
ax.set_ylabel("kmT")


# loop over the different kmL and kmT values
for kmL in kmL_vals:
    for kmT in kmT_vals:
        params = [km0L, kmL, KL, nL, kdmL, kpL, kdpL, km0T, kmT, KT, nT, kdmT, kpT, kdpT]

        # Initial conditions close to LacI high steady state
        ML0 = 36
        PL0 = 2600
        MT0 = 0
        PT0 = 0
        s0_laci = [ML0, PL0, MT0, PT0]
        
        s0_laci = [36, 2600, 0, 0]
        t_start = 0
        t_end = 10000
        t_obs = np.arange(t_start, t_end + 0.1, 1)

        # Simulate close to LacI high steady state
        s_obs_laci = odeint(sdot2, s0_laci, t_obs, args=(params,))
        sim1_PL_ss = s_obs_laci[-1, 1]  # Final PL state

        # Initial conditions close to TetR high steady state
        ML0 = 0
        PL0 = 0
        MT0 = 36
        PT0 = 2600
        s0_tetr = [ML0, PL0, MT0, PT0]
        t_start = 0
        t_end = 10000
        t_obs = np.arange(t_start, t_end + 0.1, 1)

        # Simulate close to TetR high steady state
        s_obs_tetr = odeint(sdot2, s0_tetr, t_obs, args=(params,))
        sim2_PL_ss = s_obs_tetr[-1, 1]  # Final PL state

        # Compare the steady state values
        if abs(sim1_PL_ss - sim2_PL_ss) < 0.1:
            # No bistability (same steady state)
            ax.plot(kmL, kmT, 'rx', markersize=5)
        else:
            # Bistability (different steady states)
            ax.plot(kmL, kmT, 'kx', markersize=5)

                
        # compare the steady state values reached

        # if agree within some threshold (here set to 0.1) both simulations
        # ended in the same steady state 
        if abs(sim1_PL_ss - sim2_PL_ss) < 0.1:
            # plot a red cross on the the kmL / kmT plot
            ax.plot(kmL, kmT, 'rx')
        # else if they are not the same the simulations
        # ended in different steady states (bistable)
        else:
            # plot a black cross on the the kmL / kmT plot
            ax.plot(kmL, kmT, 'kx')

ax.set_xscale('log')
ax.set_yscale('log')
```
